```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
```

```python
O = np.array([0.,0])
A = np.array([1.,2])
B = np.array([3.,4])
C = np.array([-2,1])
O,A,B,C
```

```python
plt.arrow(O[0], O[1], A[0], A[1], color="blue", head_width=0.05, head_length=0.05, )
plt.arrow(O[0], O[1], B[0], B[1], color="red", head_width=0.05, head_length=0.05, )
plt.arrow(O[0], O[1], C[0], C[1], color="green", head_width=0.05, head_length=0.05, )
```

```python
euclid = np.sqrt((A[0]-B[0])**2+(A[1]-B[1])**2)
norm = np.linalg.norm(A-B)
euclid, norm
```

```python
scalar = (A[0]**2+A[1]**2+B[0]**2+B[1]**2-((A[0]-B[0])**2+(A[1]-B[1])**2)) * 0.5
inner = np.inner(A,B)
dot = np.dot(A,B)
scalar, inner, dot
```

```python
cosine=distance.cosine(A,B)
cosine
```

```python
np.inner(A,C), 1 - distance.cosine(A,C)
```

```python
np.inner(A,A), 1 - distance.cosine(A,A)
```

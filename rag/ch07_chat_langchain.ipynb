```python
import langchain
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
import os
langchain.__version__
```

```python
with open("data/secrets/key.secret") as f:
    os.environ["OPENAI_API_KEY"] = f.read()
```

```python
chat = ChatOpenAI(model_name="gpt-4", temperature=0.7) #temperature contr√¥le le niveau de cr√©ativit√© ou d‚Äôimpr√©visibilit√© des r√©ponses g√©n√©r√©es par le mod√®le.

def start_chat():
    print("Bienvenue dans le chatbot LangChain + OpenAI ! (tapez 'exit' pour quitter)")
    while True:
        user_input = input("Vous : ")
        if user_input.lower() == "exit":
            print("Fin du chat. √Ä bient√¥t !")
            break

        # Cr√©ation du message utilisateur
        message = HumanMessage(content=user_input)

        # G√©n√©ration de la r√©ponse
        response = chat.invoke([message])
        print("Bot :", response.content)

start_chat()
```
